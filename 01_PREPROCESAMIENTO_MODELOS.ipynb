```python
#Debido a que 'Color' tiene pocos datos faltantes y 'Transmission' tambien no generaría ningún problema reemplazarlos con la moda
df_kick["Color"]=df_kick["Color"].apply(lambda x: x if not pd.isnull(x) else (df_kick["Color"].mode()[0]))
df_kick["Transmission"]=df_kick["Transmission"].apply(lambda x: x if not pd.isnull(x) else (df_kick["Transmission"].mode()[0]))

# Para 'WheelType': debido a que son pocos datos faltantes se podría reemplazar por la moda, tampoco generaría problemas la marca de una llanta
df_kick["WheelType"]=df_kick["WheelType"].apply(lambda x: x if not pd.isnull(x) else (df_kick["WheelType"].mode()[0]))

# Al ser tan pocos datos faltantes 'Size' se rellena con su moda
df_kick["Size"]=df_kick["Size"].apply(lambda x: x if not pd.isnull(x) else (df_kick["Size"].mode()[0]))
```

```python
# Para el caso de las variables MMR, que se tratan de precios se podrían reemplazar los datos faltantes con el valor medio de cada columna
# Además, no representan tantos datos y no generaría grandes problemas

df_kick["MMRAcquisitionAuctionAveragePrice"]=df_kick["MMRAcquisitionAuctionAveragePrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRAcquisitionAuctionAveragePrice"].mean()))
df_kick["MMRAcquisitionAuctionCleanPrice"]=df_kick["MMRAcquisitionAuctionCleanPrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRAcquisitionAuctionCleanPrice"].mean()))
df_kick["MMRAcquisitionRetailAveragePrice"]=df_kick["MMRAcquisitionRetailAveragePrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRAcquisitionRetailAveragePrice"].mean()))
df_kick["MMRAcquisitonRetailCleanPrice"]=df_kick["MMRAcquisitonRetailCleanPrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRAcquisitonRetailCleanPrice"].mean()))
df_kick["MMRCurrentAuctionAveragePrice"]=df_kick["MMRCurrentAuctionAveragePrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRCurrentAuctionAveragePrice"].mean()))
df_kick["MMRCurrentAuctionCleanPrice"]=df_kick["MMRCurrentAuctionCleanPrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRCurrentAuctionCleanPrice"].mean()))
df_kick["MMRCurrentRetailAveragePrice"]=df_kick["MMRCurrentRetailAveragePrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRCurrentRetailAveragePrice"].mean()))
df_kick["MMRCurrentRetailCleanPrice"]=df_kick["MMRCurrentRetailCleanPrice"].apply(lambda x: x if not pd.isnull(x) else int(df_kick["MMRCurrentRetailCleanPrice"].mean()))
```

# MODELOS

```python
from scipy import stats
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df_kick_dummies = pd.get_dummies(df_kick)
df_kick_dummies
```

```python
df_kick_dummies.columns.value_counts().sum()
```

```python
# Particion de dataframe

from sklearn.model_selection import train_test_split

X = df_kick_dummies.drop(columns=['IsBadBuy'])
y = df_kick_dummies['IsBadBuy']

# Divide el DataFrame en conjuntos de entrenamiento (40%) y prueba (60%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, random_state=42)

#Escalado

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
# Dataframe escalado

from sklearn.preprocessing import StandardScaler

# Crea una instancia de StandardScaler
scaler = StandardScaler()

# Ajusta el escalador a los datos de entrenamiento
scaler.fit(X_train)

# Transforma los datos de entrenamiento y prueba
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Ahora, X_train_scaled y X_test_scaled contienen los datos escalados

print(X_train_scaled.shape, y_train.shape)
print(X_test_scaled.shape, y_test.shape)
```

```python
# Regresion logistica

model_lr = LogisticRegression()
model_lr.fit(X_train_scaled, y_train)
```

```python
# Prediccion

y_pred = model_lr.predict(X_test_scaled)
y_pred
```

```python
# Recuento de valores unicos - Regresion logistica

unique_values, counts = np.unique(y_pred, return_counts=True)

# unique_values contiene los valores únicos, y counts contiene sus recuentos respectivos
for value, count in zip(unique_values, counts):
    print(f'Valor único: {value}, Recuento: {count}')
```

```python
# Accuracy para test regresion logistica

print("accuracy train %.2f"%model_lr.score(X_test_scaled,y_test))
tr_preds = model_lr.predict(X_test_scaled)
print("predicciones para train")
print(tr_preds)
print("ground truth para train")
print(y_test)
```

```python
# F1 score para test regresion logistica

from sklearn.metrics import f1_score

# Suposicion de predicciones y etiquetas reales
y_pred = model_lr.predict(X_test_scaled)  # Reemplazar 'clf' con modelo entrenado
y_true = y_test  # Reemplazar 'y_test' con etiquetas reales

# Calcular el puntaje F1
f1 = f1_score(y_true, y_pred)

print("Puntaje F1:", f1)
```

```python
# Random forest

from sklearn.ensemble import RandomForestClassifier

# Crear una instancia de RandomForestClassifier
clf = RandomForestClassifier(random_state=42)

# Ajustar el modelo a tus datos de entrenamiento
clf.fit(X_train_scaled, y_train)
```

```python
y_pred_rf = clf.predict(X_test_scaled)
y_pred_rf
```

```python
# Recuento de valores unicos - Random forest

unique_values, counts = np.unique(y_pred_rf, return_counts=True)

# unique_values contiene los valores únicos, y counts contiene sus recuentos respectivos
for value, count in zip(unique_values, counts):
    print(f'Valor único: {value}, Recuento: {count}')
```

```python
# Accuracy para test random forest

print("accuracy train %.2f"%clf.score(X_test_scaled,y_test))
tr_preds = clf.predict(X_test_scaled)
print("predicciones para train")
print(tr_preds)
print("ground truth para train")
print(y_test)
```

```python
# F1 score para test random forest

from sklearn.metrics import f1_score

# Suposicion de predicciones y etiquetas reales
y_pred = clf.predict(X_test_scaled)  # Reemplazar 'clf' con modelo entrenado
y_true = y_test  # Reemplazar 'y_test' con etiquetas reales

# Calcular el puntaje F1
f1 = f1_score(y_true, y_pred)

print("Puntaje F1:", f1)
```
